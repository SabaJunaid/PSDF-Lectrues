# Python Syntax and Indentation

## What is Syntax?

Syntax means the set of rules used to write programs in a programming language.

Python syntax is simple and readable.

---

## What is Indentation?

Indentation means giving spaces before writing code inside blocks like `if`, `for`, and functions.

In Python, indentation is very important because it defines the structure of the program.

````python
if 5 > 2:
    print("5 is greater than 2")
````

# Comments in Python

Comments are used to explain code.  
Python ignores comments during execution.

## Single Line Comment

Single-line comments are written using `#`.

### Example

````python
# This is a single-line comment
print("Hello World")
````

## Multi-line Comment

Multi-line comments are written using triple quotes `''' '''` or `""" """`.

### Example


````python
"""
This is a multi-line comment
in Python
"""
print("Python")
````

# Variables in Python

## What is a Variable?

A variable is used to store data or values in memory.

### Example

````python
name = "Ali"
age = 20

print(name)
print(age)
````

# Rules for Naming Variables

## Valid Rules

1. Variable name can contain:
   - Letters
   - Numbers
   - Underscore `_`

2. Variable name cannot start with a number.

3. Variable names are case-sensitive.

4. Spaces are not allowed.

5. Keywords cannot be used as variable names.

# Correct Variable Names

```python
name = "Ali"
student_age = 20
marks1 = 90
```

---

# Incorrect Variable Names

```python
1name = "Ali"      # Cannot start with number
student age = 20   # Spaces are not allowed
class = "BSCS"     # Keyword cannot be used
```

# 1. for Loop

The `for` loop is used when we know how many times we want to repeat code.

---

## Example 1



````python
for i in range(5):
    print(i)
````

````python
for i in range(1, 6):
    print(i)
````

# 2. while Loop

The `while` loop repeats as long as the condition is True.

---

## Example 1


````python
count = 1

while count <= 5:
    print(count)
    count += 1
````
