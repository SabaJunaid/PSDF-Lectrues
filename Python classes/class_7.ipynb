# 1. Regular Expressions (Regex)

Regular Expressions (Regex) are used to search, match, validate, and extract patterns from text.

Python provides the built-in re module for working with regex.

```python
import re
```

| Pattern | Meaning                     |
| ------- | --------------------------- |
| `\d`    | Any digit (0-9)             |
| `\D`    | Any non-digit               |
| `\w`    | Letters, digits, underscore |
| `\s`    | Whitespace                  |
| `.`     | Any character               |
| `+`     | One or more occurrences     |
| `*`     | Zero or more occurrences    |
| `^`     | Starts with                 |
| `$`     | Ends with                   |


```python
import re

text = "My age is 25 and my friend is 30."

numbers = re.findall(r"\d+", text)

print(numbers)
```

```python
import re

email = "student@gmail.com"

pattern = r"^[a-zA-Z0-9._]+@[a-zA-Z]+\.[a-zA-Z]+$"

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
```

# 2. List Comprehension

List Comprehension provides a shorter and cleaner way to create lists.

Traditional Way

```python
squares = []

for num in range(1, 6):
    squares.append(num ** 2)

print(squares)
```

```python
new_list = [expression for item in iterable]
```

```python
new_list = [expression for item in iterable if condition]
```

```python
squares = [num ** 2 for num in range(1, 6)]

print(squares)
```

```python
even_numbers = [num for num in range(1, 21) if num % 2 == 0]

print(even_numbers)
```

```python
names = ["ali", "ahmed", "sara"]

upper_names = [name.upper() for name in names]

print(upper_names)
```

# 3. Generators

A Generator produces values one at a time instead of storing everything in memory.

Generators use the yield keyword.

Why Use Generators?

Suppose you need 1 million numbers.

A list stores all numbers in memory.

A generator creates them only when needed.

# 4. Decorators

A Decorator is used to add extra functionality to a function without modifying its original code.

Think of it as wrapping a gift box.

The gift remains the same, but you add decoration around it.

```python
def login_required(func):

    def wrapper():
        print("Checking User Login...")
        func()

    return wrapper


@login_required
def dashboard():
    print("Dashboard Opened")


dashboard()
```

# 5. Getter and Setter

Getters and Setters are used to control access to class attributes.

They help implement Encapsulation.

```python
class Student:

    def __init__(self):
        self.age = 20

student = Student()

student.age = -5
print(student.age)
```

```python
class Student:

    def __init__(self):
        self.__age = 0

    def set_age(self, age):
        if age > 0:
            self.__age = age

    def get_age(self):
        return self.__age


student = Student()

student.set_age(22)

print(student.get_age())
```

# Four Pillars of OOP

Object-Oriented Programming is based on four major pillars.

* Encapsulation
* Abstraction
* Inheritance
* Polymorphism

## 1. Encapsulation
Definition

Encapsulation means hiding data and controlling access through methods.

Example

```python
class BankAccount:

    def __init__(self):
        self.__balance = 1000

    def get_balance(self):
        return self.__balance


account = BankAccount()

print(account.get_balance())
```

## 2. Abstraction
Definition

Abstraction means showing only necessary details and hiding implementation details.

Example

When driving a car:
You use:
✔ Steering
✔ Brake
✔ Accelerator

You do not need to know:
✖ Engine internals
✖ Fuel injection system
✖ Gear mechanisms

## 3. Inheritance
Definition

Inheritance allows one class to acquire properties and methods of another class.

```python
class Animal:

    def eat(self):
        print("Animal is Eating")


class Dog(Animal):

    def bark(self):
        print("Dog is Barking")


dog = Dog()

dog.eat()
dog.bark()
```

## 4. Polymorphism
Definition

Polymorphism means one interface, many forms.

The same method behaves differently for different objects.

```python
class Cat:

    def sound(self):
        print("Meow")


class Dog:

    def sound(self):
        print("Bark")


animals = [Cat(), Dog()]

for animal in animals:
    animal.sound()
```
