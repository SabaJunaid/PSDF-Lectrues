# Deep Dive: Python Operators

## Arithmetic Operators

| Operator | Name           | Description                                            | Code Example     | Output |
| -------- | -------------- | ------------------------------------------------------ | ---------------- | ------ |
| `+`      | Addition       | Adds two values together                               | `print(10 + 5)`  | `15`   |
| `-`      | Subtraction    | Subtracts right operand from left                      | `print(10 - 5)`  | `5`    |
| `*`      | Multiplication | Multiplies two values                                  | `print(10 * 5)`  | `50`   |
| `/`      | Division       | Divides left operand by right (always returns a float) | `print(10 / 4)`  | `2.5`  |
| `%`      | Modulus        | Returns the remainder of a division                    | `print(10 % 3)`  | `1`    |
| `**`     | Exponentiation | Raises left operand to the power of the right          | `print(2 ** 3)`  | `8`    |
| `//`     | Floor Division | Divides and rounds down to the nearest whole number    | `print(10 // 4)` | `2`    |


```python
# = (Basic Assignment)
x = 10
print(x)  # Output: 10

# += (Addition Assignment)
x += 5    # Equivalent to: x = x + 5
print(x)  # Output: 15

# -= (Subtraction Assignment)
x -= 3    # Equivalent to: x = x - 3
print(x)  # Output: 12

# *= (Multiplication Assignment)
x *= 2    # Equivalent to: x = x * 2
print(x)  # Output: 24

# /= (Division Assignment)
x /= 4    # Equivalent to: x = x / 4
print(x)  # Output: 6.0

# %= (Modulus Assignment)
x %= 4    # Equivalent to: x = x % 4
print(x)  # Output: 2.0

# //= (Floor Division Assignment)
y = 10
y //= 3   # Equivalent to: y = y // 3
print(y)  # Output: 3

# **= (Exponentiation Assignment)
z = 3
z **= 2   # Equivalent to: z = z ** 2
print(z)  # Output: 9
```

## Comparison Operators

```python
a = 10
b = 20

# == (Equal to)
print(a == b)  # Output: False

# != (Not equal to)
print(a != b)  # Output: True

# > (Greater than)
print(a > b)   # Output: False

# < (Less than)
print(a < b)   # Output: True

# >= (Greater than or equal to)
print(a >= 10) # Output: True

# <= (Less than or equal to)
print(b <= 20) # Output: True
```

## Logical Operators

```python
x = 5

# and (Returns True if BOTH conditions are true)
print(x > 3 and x < 10)  # True and True -> Output: True

# or (Returns True if AT LEAST ONE condition is true)
print(x > 10 or x < 4)   # False or False -> Output: False
print(x > 3 or x < 2)    # True or False -> Output: True

# not (Reverses the Boolean result)
print(not(x > 3))        # not(True) -> Output: False
```

## Membership Operators

```python
phrase = "Learning Python is fun"
numbers = [10, 20, 30, 40]

# in (Returns True if the sequence is found)
print("Python" in phrase)
print(50 in numbers)

# not in (Returns True if the sequence is missing)
print("Java" not in phrase)
print(20 not in numbers)
```

# String Operations, Slicing, and Functions

## Basic String Operations

```python
str1 = "Hello "
str2 = "World"

# Concatenation (Joining strings with +)
combined = str1 + " " + str2
print(combined)  # Output: Hello World

# Repetition (Repeating strings with *)
repeated = str1 * 3
print(repeated)  # Output: HelloHelloHello
```

## String Slicing

string[start:end:step]

```python
text = "PythonProgramming"

# Get a single character (Zero-indexed)
print(text[0])
# Slicing from start to end index (Excluding end)
print(text[0:6])

# Slicing without specifying the start (Defaults to index 0)
print(text[:6])

# Slicing without specifying the end (Goes to the very end)
print(text[6:])

# Negative Indexing (Counting backward from the end)
print(text[-1])


# Using Steps (Extract every second character)
print(text[0:12:2])

# Reversing a string entirely using a negative step
print(text[::-1]) # Output: gnimmargorPnohtyP
```

## Comprehensive String Functions

```python
sample = "__hello python architecture!__"

# upper() - Converts string to uppercase
print(sample.upper())

# lower() - Converts string to lowercase
print(sample.lower())

# strip() - Removes whitespace from both beginning and end
clean_sample = sample.strip("_")
right = sample.rstrip("_")
left = sample.lstrip("_")
print("right"+ right)
print("Left"+left)
print("Whole"+clean_sample)

# replace() - Replaces a phrase with another phrase
print(clean_sample.replace("architecture", "world"))

# split() - Splits string into a list based on a separator delimiter
words = clean_sample.split("python")
print(words)


# find() - Searches string for a value and returns its index position
print(clean_sample.find("python"))

# startswith() & endswith() - Return Boolean checks
print(sample.startswith("hello"))
print(sample.endswith("?"))
```

# Sets

Key Characteristics

Unordered: Items do not have a defined index order; you cannot look up elements by index.

Unchangeable Elements: You cannot modify elements directly, but you can add or remove items.

No Duplicates: Duplicate values are automatically dropped.

```python
# Initializing sets
set_a = {1, 2, 3, 4, 5}

set_b = {4, 5, 6, 7, 8}

# 1. add() - Adds a single element to the set
set_a.add(10)
print("After add:", set_a)

# 2. remove() Removes a specific element. Raises a KeyError if element is missing.
set_a.remove(10)

# 3. discard() Removes an element. Does NOT raise an error if element is missing.
set_a.discard(99) # Safe to call even though 99 is not in the set

# 4. pop() - Removes and returns a random element from the set
popped_val = set_a.pop()
print(f"Popped value: {popped_val}")

# 5. union() - Combines elements from both sets into a new set
union_set = set_a.union(set_b)
print("Union:", union_set)

# 6. intersection() - Returns a new set with elements common to both sets
intersect_set = set_a.intersection(set_b)
print("Intersection:", intersect_set)

# 7. difference() - Returns elements in set_a that are missing from set_b
diff_set = set_a.difference(set_b)
print("Difference (A - B):", diff_set)
```

```python

print("Original Dictionary:", student)

# 1. Accessing values safely using get()
# get() prevents program crashes by returning None (or a fallback value) if the key is missing
print("Access name via get():", student.get("name"))
print("Access missing key via get():", student.get("scholarship"))


# 2. keys(): Returns a view object containing all dictionary keys
print("Keys view:", student.keys())

# 3. values(): Returns a view object containing all dictionary values
print("Values view:", student.values())

# 4. items(): Returns a view object containing tuples of key-value pairs
print("Items view:", student.items())

# 5. update(): Updates existing key-value pairs or adds completely new ones
student.update({"gpa": 3.9})
student["major"] =  "Computer Science"
print("After update():", student)

# 6. pop(): Removes the specified key and returns its associated value
removed_gpa = student.pop("gpa")
print(f"Removed GPA value: {removed_gpa}")
print("Dictionary after pop():", student)

# 7. popitem(): Removes and returns the last inserted key-value pair as a tuple
last_item = student.popitem()
print(f"Removed last item via popitem(): {last_item}")
print("Dictionary after popitem():", student)

# 8. Loop iteration structure for dictionaries
print("\n--- Iterating Through Dictionary Keys & Values ---")
for key, value in student.items():
    print(f"Student Data: {key} -> Value: {value}")
```
