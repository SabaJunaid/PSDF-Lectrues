# Functions in Python
## What is a Function?

A function is a block of reusable code that performs a specific task.

Functions help us:

Avoid repeating code
Make programs cleaner and organized
Improve readability
Make debugging easier
Reuse logic multiple times

## Why Do We Use Functions?

Without functions, we may need to write the same code again and again.

Functions help to:

Reduce code duplication
Divide large programs into smaller parts
Improve maintenance
Make collaboration easier
Increase reusability
## Basic Syntax of a Function

```python
def function_name():
    # code block
    print("Hello")
```

```python
def greet():
    print("Hello Student")



greet()
```

# Parameters and Arguments
## What are Parameters?

Parameters are variables written inside the function definition.

def greet(name):

    print(name)

name is a parameter

## What are Arguments?

Arguments are actual values passed to the function when calling it.

greet("Ali")

"Ali" is an argument

```python
def student_info(name, age):
    print("Student Name:", name)
    print("Student Age:", age)

student_info("Ahmed", 21)
```

# Types of Functions in Python

There are 4 major types of functions:

No Argument, No Return Value

Argument, No Return Value

No Argument, Return Value

Argument and Return Value

```python
def hello():
    print("Hello World")

hello()
```

```python
def employee_details(name, salary):
    print("Employee Name:", name)
    print("Employee Salary:", salary)

employee_details("Ahmad", 50000)
```

```python
def university_name():
    return "FAST University"

name = university_name()
print("University:", name)
```

```python
def calculate_total(price, quantity):
    total = price * quantity
    return total

bill = calculate_total(200, 5)

print("Total Bill:", bill)
```

# Default Parameters
Explanation

Default parameters already contain a value.

```python
def greet(name="Student"):
    print("Hello", name)

greet()
greet("Ali")
```

# Scope of Variables
## Local Variable

Variable created inside function.

```python
def test(name):
    x = 10
    print(x)

print(name)
test()
```

## Global Variable

Variable created outside function.

```python
x = 100

def show():
    print(x)

show()
```

# Lambda Functions in Python
## What is Lambda Function?

A lambda function is a small anonymous function written in one line.

## Anonymous means:

Function without a name

# Why Do We Use Lambda Functions?

Lambda functions are used:

For short temporary functions
When function is needed only once

```python
lambda arguments : expression
```

```python
square = lambda x: x ** 2
print(square(5))
```

# Normal Function vs Lambda Function
Normal Function

```python
def add(a, b):
    return a + b

print(add(2, 3))
```

Lambda Function

```python
add = lambda a, b: a + b

print(add(2, 3))
```

```python
multiply = lambda a, b, c: a * b * c

print(multiply(2, 3, 4))
```

```python
upper_text = lambda text: text.upper()

print(upper_text("python"))
```

```python
first_char = lambda text: text[0]

print(first_char("Ahmed"))
```

# Lambda Function with If Condition
Syntax

```python
lambda arguments : value_if_true if condition else value_if_false
```

```python
lambda x : "Even" if x%2==0 else "odd"
```

```python
check = lambda age: "Adult" if age >= 18 else "Minor"

print(check(20))
print(check(15))
```
