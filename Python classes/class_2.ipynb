```python
# AI Recruitment Decision System

# Inputs
name = input("Enter candidate name: ")
age = int(input("Enter age: "))
cgpa = float(input("Enter CGPA: "))
skills = int(input("Enter number of skills: "))
experience = int(input("Enter years of experience: "))
expected_salary = int(input("Enter expected salary: "))
communication = int(input("Enter communication score (1-10): "))

print("\n===== Recruitment Result =====")

# Reject Conditions
if age < 18 or cgpa < 2.7 or skills < 3:
    print("Candidate Rejected")

# Fraud Detection
elif expected_salary > 400000 and experience == 0:
    print("Suspicious Candidate Detected")


elif cgpa >= 2.7 and skills >= 3:
    print("Selected as Intern")

elif cgpa >= 3.3 and skills >= 7 and communication >= 7:
    print("Selected as Software Engineer")

elif cgpa >= 3.0 and skills >= 5 and experience >= 1:
    print("Selected as Junior Developer")


else:
    print("No Suitable Position Found")
```

# FOR LOOP

Used when the number of repetitions is known.

## Basic Syntax

```python
for variable in sequence:
    code
```

```python
for i in range(11):
    print(i)
```

range(start, stop, step)

```python
for i in range(1, 10):
    print(i)
```

```python
name = "Python"

for c in name:
    print(c)
```

```python
name[0]
```

```python
name = "Python"
# len(name)
for c in range(len(name)):
    print(c)
```

# break Statement

Stops loop immediately.

```python
for i in range(10):
    print(i)
    if i == 5:
        break
```

# continue Statement

Skips current iteration.

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

# WHILE LOOP

Runs until condition becomes False.

## Syntax

```python
while condition:
    code
```

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

```python

while True:
  num = int(input("Enter a number: "))
  if num == 10:
    break
```

```python
password = ""

while password != "python123":
    password = input("Enter Password: ")

print("Access Granted")
```

Lists store multiple values.

## Lists are:

Ordered
Changeable (Mutable)
Allow duplicates

```python
numbers = [1, 2, 3]
names = ["Ali", "Sara"]
mixed = [1, "Hello", 3.5]
```

```python
numbers = [1, 2, 3]
for i in numbers:
  print(i)
```

```python
numbers = [1, 2, 3]
for i in range(len(numbers)):
  print(i)
```

```python
nums = [1,2]
nums.append(3)
```
