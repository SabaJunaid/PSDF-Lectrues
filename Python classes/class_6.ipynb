# Public, Protected, and Private Members

These are called Access Modifiers.

They control how data is accessed.

## 1. Public Members

Accessible from anywhere.

Default in Python.

Example:

```python
class Student:

    def __init__(self):
        self.name = "Ali"

s1 = Student()

print(s1.name)
```

## 2. Protected Members

Protected members use a single underscore _.

They are meant for internal use.

Syntax:

```python
class Student:

    def __init__(self):
        self._name = "Ali"

s1 = Student()

print(s1._name)
```

## 3. Private Members

Private members use double underscores __.

Syntax:

```python
class Student:

    def __init__(self):
        self.__name = "Ali"

s1 = Student()

print(s1.__name)
```

```python
class Student:

    def __init__(self, name, age, password):
        self.name = name          # Public
        self._age = age           # Protected
        self.__password = password # Private

    def show_data(self):
        print("Name:", self.name)
        print("Age:", self._age)
        print("Password:", self.__password)

s1 = Student("Ali", 20, "abc123")

print(s1.name)
print(s1._age)

s1.show_data()
```
