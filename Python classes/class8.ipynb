# NumPy in Python
## What is NumPy?

NumPy (Numerical Python) is a Python library used for working with numerical data efficiently.

It provides a special data structure called an ndarray (N-dimensional array) which is much faster and more memory-efficient than Python lists when performing mathematical operations.

## Why Do We Use NumPy?

Suppose you have thousands or millions of numbers and want to perform calculations on them.

```python
import numpy as np
import time

# Create 10 million numbers
numbers_list = list(range(10_000_000))
numbers_array = np.array(numbers_list)

# -----------------------------
# Python List Operation
# -----------------------------
start = time.time()

result_list = [x * 2 for x in numbers_list]

end = time.time()

print(f"List Time: {end - start:.4f} seconds")

# -----------------------------
# NumPy Operation
# -----------------------------
start = time.time()

result_array = numbers_array * 2

end = time.time()

print(f"NumPy Time: {end - start:.4f} seconds")
```
